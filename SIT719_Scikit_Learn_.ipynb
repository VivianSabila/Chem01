Before we go to choosing and traing a model, let's learn something called pipelines. Now, a basic pipeline is composed of different parts i.ea preprocessing step and a final predictor. In Scikit however, pre-processors and transfomers follow the same API as the estimator objects where, the trenaformer objects don't have a predict method rather a tranform method that gives a newly transformed sampl matrix

Since they have the same API as estimator objects, they are combines together to form a pipeline. This can be fitted and used for prediction with fit and predict.

```python
# let's compute the accuracy scores of a pipeline on the  test data
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#load data and split into train and test sets
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#create a pipeline object
pipe = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
#fitting the pipeline
pipe.fit(X_train, y_train)

#making predictions
predictions=pipe.predict(X_test)

#Test for accuracy
print("Accuracy Score:",accuracy_score(y_test,predictions))
```

```python
#Read the dataset from github and store it in Pandas Dataframe
link_to_data = 'https://raw.githubusercontent.com/SIT719/2020-S2/master/data/payment_fraud.csv'
dfp = pd.read_csv(link_to_data, on_bad_lines='skip')
#on_bad_lines{‘error’, ‘warn’, ‘skip’} or Callable, default ‘error’ ; 'skip', skip bad lines without raising or warning when they are encountered.
```

```python
#Display the pandas dataframe
dfp
```

Now to understand our data we check for missing values and remove any missing row
s followed by converting categorical features to numerical using dummies command
as shown below

```python
#check missing values
missing_values=dfp.isnull().sum()
print(missing_values)
```

```python
# pandas.get_dummies
#pandas.get_dummies(data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)[source]
#Convert categorical variable into dummy/indicator variables.; one-hot encoding
df = pd.get_dummies(dfp, columns=['paymentMethod'])
```

```python
df
#three new columns have been added to the table—paymentMethod_creditcard, paymentMethod_paypal, and paymentMethod_storecredit:
```

All done now we focus on the scikit workflow and start off by splitting the data set into training and testing datasets. Here we will use the scikit learn train and test command

```python
#import skikit learns train test split
from sklearn.model_selection import train_test_split
```

```python
#Split arrays or matrices into random train and test subsets.
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('label', axis=1), df['label'],
    test_size=0.33, random_state=17)
```

```python
X_train
```

```python
y_train
```

Having split the model, we build a model using Scikit to classify or predict outcomes. With this we will need to import linear model and initialize the function object by using the constructor and finally feeding the x-train and y-train into the fit function so as to train the model

```python
#Import sklearn.linear_model.LogisticRegression class
from sklearn.linear_model import LogisticRegression
#initialize the LogisticRegression object by invoking the constructor
clf = LogisticRegression()
# Feed X_train and y_train (i.e., the training set) into the fit() function, resulting in a trained classifier model, which is stored in the clf object.
clf.fit(X_train, y_train)
```

Next up we want to test our trained model and making a prediction using the y-pred command and measure for accuracy.

```python
#Testing the model: we are feeding X_test into the classier object's predict() function.
y_pred = clf.predict(X_test)
#observe y_pred
y_pred
```

```python
#Import sklearn.metrics.accuracy_score class
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred, y_test))
```

```python
#Compute confusion matrix to evaluate the accuracy of a classification.
#    True Positive: Correctly predicts 'positive' when the transaction is fradulent
#    True Negative: Correctly predicts 'negative' when the transaction is not fradulent
#    False Positive: Incorrectly predicts 'positive' when the transaction is not fradulent
#    False Negative: Incorrectly predicts 'negative' when the transaction is actually fradulent
#    ------------------
#  | [[  TN,   FP ]
#  |  [  FN,   TP]]
#    ------------------
print(confusion_matrix(y_test, y_pred))
```
